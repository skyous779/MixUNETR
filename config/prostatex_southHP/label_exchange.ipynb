```python
import os
import nibabel as nib
import numpy as np


```

```python
def convert_labels(input_folder, output_folder):
    # 创建输出文件夹
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # 遍历输入文件夹中的所有文件
    for filename in os.listdir(input_folder):
        if filename.endswith('.nii.gz'):
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)
            
            # 加载 NIfTI 文件
            img = nib.load(input_path)
            data = img.get_fdata()
            
            # 将标签值进行转换
            data[data == 2] = 1
            data[data == 3] = 2
            
            # 创建新的 NIfTI 对象并保存
            new_img = nib.Nifti1Image(data, img.affine, img.header)
            nib.save(new_img, output_path)
            print(f"Converted {filename} and saved to {output_path}")

# 指定输入和输出文件夹的路径
input_folder = "/home/data1/skyous/prostatex_southHP/southHP/old_labels"
output_folder = "/home/data1/skyous/prostatex_southHP/southHP/labels"

# 调用函数进行转换
convert_labels(input_folder, output_folder)
```
